```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.csv")
df
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
all_na_cols = df.columns[df.isna().all()].tolist()
df = df.drop(columns=all_na_cols)
```

```python
numeric_cols = df.select_dtypes(include=["number"]).columns.tolist()
categorical_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
```

```python
for col in categorical_cols:
    if df[col].isna().any():
        mode_val = df[col].mode()
        if not mode_val.empty:
            df[col] = df[col].fillna(mode_val.iloc[0])
        else:
            df[col] = df[col].fillna("Unknown")
```

```python
df_processed = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
output_path = "/content/INX_Future_Inc_Employee_Performance_processed.csv"
df_processed.to_csv(output_path, index=False)
```
